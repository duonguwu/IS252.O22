```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import datasets,tree,metrics
```

```python
digits = datasets.load_digits()
```

```python
_, axes =plt.subplots(1,4)
images_and_labels = list(zip(digits.images,digits.target))
for ax,(image,label) in zip(axes,images_and_labels[:4]):
    ax.set_axis_off()
    ax.imshow(image,cmap=plt.cm.gray_r,interpolation='nearest')
    ax.set_title('Trianing:%i'% label)
plt.show()
```

```python
# ma trận 8x8 thành ma trận 1x64
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))
```

```python
# Sử dụng cây ID3 entropy
classifier = tree.DecisionTreeClassifier(criterion="entropy",random_state=33)
```

```python
x_train,x_test,y_train,y_test = train_test_split(data,digits.target,test_size=0.2,shuffle=False)
```

```python
classifier.fit(x_train, y_train)
```

```python
predicted = classifier.predict(x_test)
```

```python
_, axes =plt.subplots(1,4)
images_and_predictions = list(zip(digits.images[n_samples//2:],predicted))
for ax,(image,prediction) in zip(axes,images_and_predictions[:4]):
    ax.set_axis_off()
    ax.imshow(image,cmap=plt.cm.gray_r,interpolation='nearest')
    ax.set_title('Prediction:%i'% prediction)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

print("Classificaion report for classifier %s:\n%s\n"
      %(classifier, metrics.classification_report(y_test, predicted)))

cm = confusion_matrix(y_test, predicted)  # Tạo Confusion Matrix
disp = ConfusionMatrixDisplay(cm)  # Tạo đối tượng hiển thị
disp.plot()  # Vẽ Confusion Matrix
plt.suptitle("Confusion Matrix")  # Đặt tiêu đề bằng plt.suptitle
plt.show()  # Hiển thị Confusion Matrix

print("Confusion matrix:\n", cm)  # In Confusion Matrix dưới dạng văn bản
```
